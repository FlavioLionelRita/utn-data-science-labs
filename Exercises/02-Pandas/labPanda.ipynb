````python
"""Importar los datos de los archivos csv"""
import pandas
courses = pandas.read_csv('Courses.csv', sep=',' , encoding='ISO-8859-1')
teachers = pandas.read_csv('Teachers.csv', sep=',' , encoding='ISO-8859-1')
students = pandas.read_csv('Students.csv', sep=',' , encoding='ISO-8859-1')
students_courses = pandas.read_csv('Students_Courses.csv', sep=',' , encoding='ISO-8859-1')
````

````python
"""Queries"""
import datetime

print('Cuántos cursos diferentes hay:')
result = courses['Curso'].unique()
print(result)

print('Cuántos alumnos hay en cada curso:')    
alumnos_por_curso =  students_courses.groupby('Curso ID')['EstudianteID'].agg(Count='count')
merged = pandas.merge(alumnos_por_curso, courses, on='Curso ID')
result = merged[['Curso', 'Count']]
print(result)    

print('Determinar a partir de la respuesta anterior cuál es el curso más popular:')
maxima_cantidad_de_alumnos_en_un_curso = alumnos_por_curso['Count'].max()
curso_id_con_mas_alumnos = alumnos_por_curso.loc[alumnos_por_curso['Count'] == maxima_cantidad_de_alumnos_en_un_curso].iloc[0]
curso_con_mas_alumnos = courses.loc[courses['Curso ID'] == curso_id_con_mas_alumnos.name]
print(curso_con_mas_alumnos)

print('De ese curso más popular extraer la lista de alumnos:')
alumnos_del_curso_mas_popular = students_courses.loc[students_courses['Curso ID'] == curso_id_con_mas_alumnos.name]
print(alumnos_del_curso_mas_popular)

print('Definir en función del año de ingreso cuántos años hace que esta cada estudiante en la universidad:')
student_formatead = []
current = datetime.datetime.now()
for student in students.to_dict('records'):
    new_student = {}
    fecha_ingreso = student['FechaIngreso']
    fecha = datetime.datetime.strptime(fecha_ingreso, '%m/%d/%Y')
    new_student['Years'] =  (current - fecha).days // 365
    new_student['Id'] = student['EstudianteID']
    new_student['Name'] = student['Name']
    student_formatead.append(new_student)
result = [{'Name': obj['Name'], 'Years': obj['Years']} for obj in student_formatead]    
for item in result:
    print(item)
````
